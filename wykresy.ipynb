```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('porownania.csv', sep=';')
df
```

```python
df=df.set_index('Length')
df=df[['Bubble Sort', 'Insertion Sort', 'Heap Sort', 'Quick Sort']]
df
```

```python
kolory = {
    'Bubble Sort': 'red',
    'Insertion Sort': 'green',    
    'Heap Sort': 'blue',
    'Quick Sort': '#77B5AB'
    
}
aktualne_kolory = [kolory[kolumna] for kolumna in df.columns]
```

```python
fig, ax=plt.subplots(figsize=(16,9))

ax.plot(df[['Bubble Sort', 'Insertion Sort']])

kolumny = ['Bubble Sort', 'Insertion Sort']
for i, col in enumerate(kolumny):
    ax.get_lines()[i].set_color(kolory[col])
    
plt.title('Porównanie liczby wykonanych porównań (BubbleSort vs InsertionSort)')
plt.ylabel('Liczba porównań')
plt.xlabel('Długość sortowanej listy')

plt.xlim(0,1030)
plt.ylim(0,520000)

plt.xticks(list(np.arange(10, 1001, 50)) + [1000])
plt.grid(True)

ax.legend(['Bubble Sort', 'Insertion Sort'], loc='upper left')
plt.savefig('BubIns.png')
```

```python
fig, ax=plt.subplots(figsize=(16,9))

plt.plot(df[['Insertion Sort', 'Heap Sort']])

kolumny = ['Insertion Sort', 'Heap Sort']
for i, col in enumerate(kolumny):
    ax.get_lines()[i].set_color(kolory[col])

plt.title('Porównanie liczby wykonanych porównań (Heapsort vs QuickSort)')
plt.ylabel('Liczba porównań')
plt.xlabel('Długość sortowanej listy')

plt.xlim(0,1030)
plt.ylim(0,270000)

plt.xticks(list(np.arange(10, 1001, 50)) + [1000])
plt.grid(True)

ax.legend(['Insertion Sort', 'Heap Sort'], loc='upper left')
plt.savefig('InsHeap.png')
```

```python
fig, ax=plt.subplots(figsize=(16,9))

plt.plot(df[['Heap Sort', 'Quick Sort']])

kolumny = ['Heap Sort', 'Quick Sort']
for i, col in enumerate(kolumny):
    ax.get_lines()[i].set_color(kolory[col])

plt.title('Porównanie liczby wykonanych porównań  (Heapsort vs QuickSort)')
plt.ylabel('Liczba porównań')
plt.xlabel('Długość sortowanej listy')

plt.xlim(0,1030)
plt.ylim(0,20000)

plt.xticks(list(np.arange(10, 1001, 50)) + [1000])
plt.grid(True)

ax.legend(['Heap Sort', 'Quick Sort'], loc='upper left')
plt.savefig('HeaQui.png')
```

```python
fig, ax=plt.subplots(figsize=(16,9))

df.plot(ax=ax, color=aktualne_kolory)
plt.title('Porównanie liczby wykonanych porównań (BubbleSort vs InsertionSort vs HeapSort vs QuickSort)')
plt.ylabel('Liczba porównań')
plt.xlabel('Długość sortowanej listy')

plt.xlim(0,1030)
plt.ylim(0,520000)

plt.xticks(list(np.arange(10, 1001, 50)) + [1000])
plt.grid(True)

ax.legend(df.columns)
plt.savefig('All.png')
```
